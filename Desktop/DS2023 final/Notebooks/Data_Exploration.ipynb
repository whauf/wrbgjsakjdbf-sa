```python

# Number 2 
# ==========================
# COMPUTE LORENZ CURVE
# ==========================

# Sort NIL values ascending
values = np.sort(df["NIL_Value"].values)

# Cumulative share of athletes (x-axis)
cum_pop = np.linspace(0, 1, len(values))

# Cumulative share of NIL value (y-axis)
cum_vals = np.cumsum(values) / values.sum()

# Prepend zero for the origin
cum_pop = np.insert(cum_pop, 0, 0)
cum_vals = np.insert(cum_vals, 0, 0)

# ==========================
# COMPUTE GINI COEFFICIENT
# ==========================
gini = 1 - 2 * np.trapz(cum_vals, cum_pop)

# ==========================
# PLOT LORENZ CURVE
# ==========================
plt.figure(figsize=(10, 7))

# Perfect equality line
plt.plot(cum_pop, cum_pop, color="gray", linestyle="--", label="Perfect Equality")

# Actual Lorenz curve
plt.plot(cum_pop, cum_vals, color="#4C72B0", linewidth=3, label="Lorenz Curve")

# Fill inequality area
plt.fill_between(cum_pop, cum_vals, cum_pop, color="#4C72B0", alpha=0.25)

# Title & labels
plt.title("Lorenz Curve of NIL Distribution\n(Gini Coefficient = {:.3f})".format(gini),
          fontsize=18, weight="bold")
plt.xlabel("Cumulative Share of Players", fontsize=14)
plt.ylabel("Cumulative Share of NIL Value", fontsize=14)

# Annotation explaining inequality
plt.annotate(
    "Most NIL value is concentrated among a\nsmall group of top players.",
    xy=(0.92, 0.78),               # <-- arrow points here (adjust for your curve)
    xytext=(0.55, 0.55),           # <-- text placed here
    arrowprops=dict(arrowstyle="->", lw=2),
    fontsize=13
)


# Source information
plt.text(
    0.0, -0.12,
    "Source: ON3 NIL Valuations | Note: Estimated values, not contractual payouts.",
    fontsize=11,
    transform=plt.gca().transAxes
)

plt.legend()
sns.despine()
plt.tight_layout()
plt.show()
```

```python
# Aggregate NIL by conference
conf_share = df.groupby("Conference")["NIL_Value"].sum().sort_values(ascending=False)

# Colors
colors = sns.color_palette("tab10", len(conf_share))

plt.figure(figsize=(12, 10))

# Donut chart
wedges, texts = plt.pie(
    conf_share,
    labels=None,
    startangle=90,
    colors=colors,
    wedgeprops=dict(width=0.35, edgecolor="white")
)

# Create circle for donut hole
centre_circle = plt.Circle((0,0), 0.55, color='white')
plt.gca().add_artist(centre_circle)

# Title moved upward to avoid label collision
plt.title(
    "Share of Total NIL Value by Conference\n(Structural Inequality Across Conferences)",
    fontsize=19,
    weight="bold",
    pad=40      # <-- PUSH TITLE UPWARD
)

# Add extra room above chart
plt.subplots_adjust(top=0.82)

# Compute slice angles
angles = [(w.theta2 + w.theta1) / 2 for w in wedges]
angles_rad = np.deg2rad(angles)

# Position for labels
base_label_radius = 1.45  # slightly further out

for i, ((conf, val), ang) in enumerate(zip(conf_share.items(), angles_rad)):
    percent = val / conf_share.sum() * 100
    label = f"{conf}\n{percent:.1f}% (${val:,.0f})"
    
    # Base coordinates
    x = np.cos(ang) * base_label_radius
    y = np.sin(ang) * base_label_radius

    # ---- FIX: lower top labels so they don't touch the title ----
    if y > 0.7:       # label is near the top
        y -= 0.35     # move label down more
    elif y > 0.5:     # slightly upper
        y -= 0.18

    # Arrow anchor
    x_arrow = np.cos(ang) * 0.9
    y_arrow = np.sin(ang) * 0.9

    # Label text
    plt.text(
        x, y, label,
        ha="center",
        va="center",
        fontsize=12,
        weight="bold",
        bbox=dict(facecolor='white', edgecolor='none', alpha=0.8, boxstyle='round,pad=0.3')
    )

    # Leader line
    plt.annotate(
        "", xy=(x_arrow, y_arrow), xytext=(x, y),
        arrowprops=dict(arrowstyle="-", lw=1.5, color="gray")
    )

# Footer note
plt.text(
    0, -1.35,
    "Source: ON3 NIL Valuations | Note: NIL values are estimates and subject to change.",
    ha="center",
    fontsize=12
)

plt.axis("equal")
plt.tight_layout()
plt.show()
```
